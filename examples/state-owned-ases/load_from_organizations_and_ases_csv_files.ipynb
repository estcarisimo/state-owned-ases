# 1. Imports

```python
import pandas as pd
```

# 2. Load data

```python
orgs = pd.read_csv("../../data/state-owned-ases/csv/state_owned_ases_table_organizations.csv", header="infer")
ases = pd.read_csv("../../data/state-owned-ases/csv/state_owned_ases_table_ases.csv", header="infer")
```

# 3. Get insights of the dataset

```python
orgs.head()
```

```python
ases.head()
```

```python
print(f"The organization Data Frame cointains {orgs.shape[0]} rows and {orgs.shape[1]} columns")
print(f"The ASes Data Frame cointains {ases.shape[0]} rows and {ases.shape[1]} columns")
```

# 4. Joining both Data Frame into a single one

```python
state_owned_ases = orgs.join(
    ases.set_index("org_id"),
    on="org_id",
    how='left',
    lsuffix='_left',
    rsuffix='_right'
)
```
